```python
conn_gen = df.iloc[:,:8]
travaux_specifique = df.iloc[:,8:10] 
influence_reconn = df.iloc[:,10:15] 
reception_crtiq = df.iloc[:,15:18] 
sources_info = df.iloc[:,18:19] 
info_sur_part = df.iloc[:,19:25]
```

### Imputation des variables

```python
def imputation(df1):
    
    ########################## conn_gen $$ travaux_specifique ###########################
    
    for col in [i for i in conn_gen.columns if i != conn_gen.columns[6]]:
        df1[col] = df1[col].fillna(df1[col].mode()[0])
    df1[conn_gen.columns[6]] = df1[conn_gen.columns[6]].fillna("Non")
    df1[travaux_specifique.columns[0]] = df1[travaux_specifique.columns[0]].fillna("Aucune")
    df1[travaux_specifique.columns[1]]=df1[travaux_specifique.columns[1]].fillna("Non")
    
    ################################### influence_reconn ################################
    
    liste = list(influence_reconn.columns[3:5])            
    liste.append(influence_reconn.columns[1])
    for col in [i for i in influence_reconn.columns if i not in liste]:
        df1[col] = df1[col].fillna(df1[col].mode()[0])
    for col in liste:
        df1[col] = df1[col].fillna("Non")
     
    ########################## reception_crtiq $$ sources_info ###########################
    
    for col in reception_crtiq.columns[0::2]:
        df1[col] = df1[col].fillna(df1[col].mode()[0])
    df1[reception_crtiq.columns[1]] =  df1[reception_crtiq.columns[1]].fillna("Unknow")   
    df1[sources_info.columns[0]] = df1[sources_info.columns[0]].fillna(df1[sources_info.columns[0]].mode()[0])
    
    ###################################  info_sur_part  ##################################
    
    for col in info_sur_part.columns[0:3]:
        df1[col] = df1[col].fillna(df1[col].mode()[0])
    for col in info_sur_part.columns[3:5]:
        df1[col] = df1[col].fillna(df1[col].mean())     
    df1[info_sur_part.columns[5]] = df1[info_sur_part.columns[5]].fillna("Je ne souhaite pas poursuivre")
    
    ######################################################################################
    
    #df1[question_ouvertes.columns[0]] = df1[question_ouvertes.columns[0]].fillna("Pas de commentaire")
    #df1[question_ouvertes.columns[1]] = df1[question_ouvertes.columns[1]].fillna("Je ne souhaite pas poursuivre")
    
imputation(df) 
```

## Traitement des variables 

```python
from unidecode import unidecode

############################### Fonction qui affiche les modalites de la colonnes voulues ##########################

def mod(df1, col):
    val = df1[col].value_counts().values
    data = pd.DataFrame({"modalites":df1[col].value_counts().index,"nombre":df1[col].value_counts().values,"proportions":val/val.sum()})
    #print(f"{col}")
    return (data)

def Autre_1(don,df1, col):
    for i in range(len(df1[col])):
        for j in range(len(don["modalites"])):
            if df1.loc[i,col] == don.loc[j,"modalites"] and don.loc[j,"nombre"] ==1:
                df1.loc[i,col] = "Autre"
    #return df1[col]

def Autre_5(don,df1,col):
    for i in range(len(df1[col])):
        for j in range(len(don["modalites"])):
            if df1.loc[i,col] == don.loc[j,"modalites"] and don.loc[j,"nombre"] <=5:
                df1.loc[i,col] = "Autre" 
    #return df1[col]
    
################################  Traitement de la variable Pays de Naissance ########################################

donn = mod(df,"Pays_naiss")
Autre_1(donn,df,"Pays_naiss")
# for i in range(len(df["Pays_naiss"])):
#     for j in range(len(donn["modalites"])):
#         if df.loc[i,"Pays_naiss"] == donn.loc[j,"modalites"] and donn.loc[j,"nombre"] ==1:
#             df.loc[i,"Pays_naiss"] = "Autre" 


################################  Traitement de la variable Qui_est_ce ###############################################

donn = mod(df,"Qui_est_ce")
Autre_5(donn,df,"Qui_est_ce")
# for i in range(len(df["Qui_est_ce"])):
#     for j in range(len(donn["modalites"])):
#         if df.loc[i,"Qui_est_ce"] == donn.loc[j,"modalites"] and donn.loc[j,"nombre"] <=5:
#             df.loc[i,"Qui_est_ce"] = "Autre" 
            
            
###################################### Traitement de la  variable Citer_oeuvre #######################################
  
for i in range(len(df)):
    df.loc[i,"Citer_oeuvre"] = df.loc[i,"Citer_oeuvre"].lower()
    if "negre" in df.loc[i,"Citer_oeuvre"]:
        df.loc[i,"Citer_oeuvre"] = "nation negre et culture"
    if "civilisation" in df.loc[i,"Citer_oeuvre"]:
        df.loc[i,"Citer_oeuvre"] = "civilisation et barbarie"
    if "afrique" in df.loc[i,"Citer_oeuvre"]:
        df.loc[i,"Citer_oeuvre"] = "afrique noire precoloniale"
    mots = (df.loc[i,"Citer_oeuvre"]).split()
    nouveau = ""
    for mot in mots:
        if mot.endswith('s'):
            mot = mot[:-1]
        nouveau += mot + " "
    df.loc[i,"Citer_oeuvre"] = unidecode(nouveau.strip())
    
donn = mod(df,"Citer_oeuvre")
Autre_1(donn,df,"Citer_oeuvre")
# for i in range(len(df["Citer_oeuvre"])):
#     for j in range(len(donn["modalites"])):
#         if df.loc[i,"Citer_oeuvre"] == donn.loc[j,"modalites"] and donn.loc[j,"nombre"] ==1:
#             df.loc[i,"Citer_oeuvre"] = "Autre"     
    
#################################### Traitement de la vaiable Oeuvres_lues ########################################### 

for i in range(len(df)):
    df.loc[i,"Oeuvres_lues"] = df.loc[i,"Oeuvres_lues"].lower()
    if "nations" in df.loc[i,"Oeuvres_lues"]:
        df.loc[i,"Oeuvres_lues"] = "nation negre et culture"
    if "civilisation" in df.loc[i,"Oeuvres_lues"]:
        df.loc[i,"Oeuvres_lues"] = "civilisation et barbarie"
    if "culturelle" in df.loc[i,"Oeuvres_lues"]:
        df.loc[i,"Oeuvres_lues"] = "l'unité culturelle de l'afrique noire,1959"


#################################### Traitement de la vaiable Nommer_influencer ######################################

for i in range(len(df)):
    df.loc[i,"Nommer_influencer"] = df.loc[i,"Nommer_influencer"].lower()
    df.loc[i,"Nommer_influencer"] = unidecode(df.loc[i,"Nommer_influencer"].strip())
    if "theophile" in df.loc[i,"Nommer_influencer"] or "theodore" in df.loc[i,"Nommer_influencer"] or "theplophile" in df.loc[i,"Nommer_influencer"] or "teophile" in df.loc[i,"Nommer_influencer"]:
        df.loc[i,"Nommer_influencer"] = "theophile obenga"
    if "panafricanisme" in df.loc[i,"Nommer_influencer"]:
        df.loc[i,"Nommer_influencer"] = "le panafricanisme"
    if "joseph" in df.loc[i,"Nommer_influencer"] or "ki" in df.loc[i,"Nommer_influencer"] or "kizerbo" in df.loc[i,"Nommer_influencer"] :
        df.loc[i,"Nommer_influencer"] = "joseph ki zerbo"
    if "aime" in df.loc[i,"Nommer_influencer"] or "aimee" in df.loc[i,"Nommer_influencer"] or "kizerbo" in df.loc[i,"Nommer_influencer"] :
        df.loc[i,"Nommer_influencer"] = "aime cesaire"
    if "omotounde" in df.loc[i,"Nommer_influencer"] or "kalala" in df.loc[i,"Nommer_influencer"] or "omotunde" in df.loc[i,"Nommer_influencer"]:
        df.loc[i,"Nommer_influencer"] = "omotounde kalala"
    if "amadou" in df.loc[i,"Nommer_influencer"] or "krouma" in df.loc[i,"Nommer_influencer"] :
        df.loc[i,"Nommer_influencer"] = "amadou kourouma"
    if "kemi" in df.loc[i,"Nommer_influencer"] or "kemite" in df.loc[i,"Nommer_influencer"]  or "keli" in df.loc[i,"Nommer_influencer"] :
        df.loc[i,"Nommer_influencer"] = "kemi seba"
        
donn = mod(df,"Nommer_influencer")
Autre_1(donn,df,"Nommer_influencer")
# for i in range(len(df["Nommer_influencer"])):
#     for j in range(len(donn["modalites"])):
#         if df.loc[i,"Nommer_influencer"] == donn.loc[j,"modalites"] and donn.loc[j,"nombre"] == 1:
#             df.loc[i,"Nommer_influencer"] = "Autre"  


################################## Traitement de la vaiable Comment_est_il_percu #####################################

donn = mod(df,"Comment_est_il_percu")
Autre_5(donn,df,"Comment_est_il_percu")

################################## Traitement de la vaiable Citer_enseignant #########################################

for i in range(len(df)):
    if df.loc[i,"Citer_enseignant"] ==0 or "je" in df.loc[i,"Citer_enseignant"]:
        df.loc[i,"Citer_enseignant"] = "non"
    df.loc[i,"Citer_enseignant"] = df.loc[i,"Citer_enseignant"].lower() 
    df.loc[i,"Citer_enseignant"] = unidecode(df.loc[i,"Citer_enseignant"].strip())
    if "theophile" in df.loc[i,"Citer_enseignant"]: 
        df.loc[i,"Citer_enseignant"] = "theophile obenga"
    if "marcel" in df.loc[i,"Citer_enseignant"]:
        df.loc[i,"Citer_enseignant"] = "marcel griaule"
    if "gaston" in df.loc[i,"Citer_enseignant"]:
        df.loc[i,"Citer_enseignant"] = "gaston bachelard"
    if "julio" in df.loc[i,"Citer_enseignant"] or "joliot" in df.loc[i,"Citer_enseignant"]: 
        df.loc[i,"Citer_enseignant"] = "pierre marie curie"
    if "marie" in df.loc[i,"Citer_enseignant"]:
        df.loc[i,"Citer_enseignant"] = "pierre marie curie"
donn = mod(df,"Citer_enseignant")
Autre_1(donn,df,"Citer_enseignant")

################################## Traitement de la vaiable Intellect_qui_a_soutenu ##################################

for i in range(len(df)):
    df.loc[i,"Intellect_qui_a_soutenu"] = df.loc[i,"Intellect_qui_a_soutenu"].lower()
    df.loc[i,"Intellect_qui_a_soutenu"] = unidecode(df.loc[i,"Intellect_qui_a_soutenu"].strip())
    if "senghor" in df.loc[i,"Intellect_qui_a_soutenu"]:
        df.loc[i,"Intellect_qui_a_soutenu"] = "leopold sedar senghor"
    if "obenga" in df.loc[i,"Intellect_qui_a_soutenu"] or "obenga;" in df.loc[i,"Intellect_qui_a_soutenu"] or "obenga," in df.loc[i,"Intellect_qui_a_soutenu"] or "obinga," in df.loc[i,"Intellect_qui_a_soutenu"]  or "obing" in df.loc[i,"Intellect_qui_a_soutenu"]:
        df.loc[i,"Intellect_qui_a_soutenu"] = "theophile obenga"
    if "joseph" in df.loc[i,"Intellect_qui_a_soutenu"]:
        df.loc[i,"Intellect_qui_a_soutenu"] = "joseph ki zerbo"
    if "aime" in df.loc[i,"Intellect_qui_a_soutenu"]:
        df.loc[i,"Intellect_qui_a_soutenu"] = "aime cesaire"

donn = mod(df,"Intellect_qui_a_soutenu")
Autre_1(donn,df,"Intellect_qui_a_soutenu")

################################## Traitement de la vaiable Intellect_qui_a_soutenu ###################################

donn = mod(df,"Critiques_formulées")
Autre_1(donn,df,"Critiques_formulées")

#################################### Traitement de la vaiable Pays_naiss_part #########################################

donn = mod(df,"Pays_naiss_part")
Autre_5(donn,df,"Pays_naiss_part")

####################################### Traitement de la vaiable Pays_resid_part ######################################

donn = mod(df,"Pays_resid_part")
Autre_5(donn,df,"Pays_resid_part")

########################################## Traitement de la vaiable Age_part #########################################

for i in range(len(df)):
    df.loc[i,"Age_part"] = abs(df.loc[i,"Age_part"])
    if df.loc[i,"Age_part"] <=15: 
        df.loc[i,"Age_part"] = "enfant"
    elif df.loc[i,"Age_part"] >15 and df.loc[i,"Age_part"] <36:
        df.loc[i,"Age_part"] = "jeune"
    else:
        df.loc[i,"Age_part"] = "adulte"
        
        
########################################### Traitement de la vaiable Age_part #########################################

for i in range(len(df)):
    df.loc[i,"Age_entendu_parler_part"] = abs(df.loc[i,"Age_entendu_parler_part"])
    if df.loc[i,"Age_entendu_parler_part"] <=15: 
        df.loc[i,"Age_entendu_parler_part"] = "enfant"
    elif df.loc[i,"Age_entendu_parler_part"] >15 and df.loc[i,"Age_entendu_parler_part"] <36:
        df.loc[i,"Age_entendu_parler_part"] = "jeune"
    else:
        df.loc[i,"Age_entendu_parler_part"] = "adulte"        

######################################################################################################################
```
